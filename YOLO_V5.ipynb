```python
for label in labels:
  image_folder_path = f'/content/Dataset/dacon_data/{label}'

  # 폴더에 있는 파일 목록을 가져옵니다.
  image_files = os.listdir(image_folder_path)
  jpg_list = []
  txt_list = []

  # 파일 목록을 반복합니다.
  for image_file in image_files:

      # 파일 이름을 가져옵니다.
      file_name = os.path.basename(image_file)

      # 파일 이름이 지정된 이름과 일치하는지 확인합니다.
      base_name, extend = os.path.splitext(file_name)[0], os.path.splitext(file_name)[1]
      if extend == '.png':
        jpg_list.append(base_name)
      if extend == '.txt':
        txt_list.append(base_name)

  check = set(jpg_list) - set(txt_list)
  print(check)
```

```python
import shutil
```

```python
# Get a list of all the files in the source directory
for label in labels:
  path = f'/content/Dataset/copy_dataset/train/{label}'
  file_list = os.listdir(path)

  source_dir = path
  destination_dir_img = f'/content/Dataset/images/train'
  destination_dir_label = f'/content/Dataset/labels/train'
  
  # Iterate through each file in the source directory
  for file_name in file_list:
      # Check if the file is an image file
      if file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
          # Construct the full file paths for the source and destination files
          source_file = os.path.join(source_dir, file_name)
          destination_file = os.path.join(destination_dir_img, file_name)
          
          # Copy the file from the source directory to the destination directory
          shutil.move(source_file, destination_file)
          
          # Print a message indicating that the file has been copied
          print(f"Copied {file_name} to {destination_dir_img}")
      elif file_name.lower().endswith(('.txt')):
          # Construct the full file paths for the source and destination files
          source_file = os.path.join(source_dir, file_name)
          destination_file = os.path.join(destination_dir_label, file_name)
          
          # Copy the file from the source directory to the destination directory
          shutil.move(source_file, destination_file)
          
          # Print a message indicating that the file has been copied
          print(f"Copied {file_name} to {destination_dir_label}")
```

```python
## yolov5 폴더 requirements.txt 수정 필요
## setuptools<=64.0.2

temp_str = 'setuptools<=64.0.2\n'

f = open('/content/yolov5/requirements.txt', 'r')
f_str = f.readlines()
f.close()

f2 = open('/content/yolov5/requirements.txt', 'w')

for idx, val in enumerate(f_str) :
    if 'setuptools' in val :
        idx_v = idx
        f_str.remove(val)
        f_str.insert(idx_v, temp_str)

for val in f_str :
    f2.write(val)

f2.close() 
```
